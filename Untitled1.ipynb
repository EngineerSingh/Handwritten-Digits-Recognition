```python
def open_file(w):
    file = filedialog.askopenfilename(filetypes = (("jpg","*.jpg"),("all files","*.*")))
    if file:
        w.delete("all")
    #path = r"C:/Users/Engineer Singh/Pictures/Screenshots/Screenshot (1).png"
    print(file)
    cv_img = cv2.imread(file)

    # Get the image dimensions (OpenCV stores image data as NumPy ndarray)
    height, width, no_channels = cv_img.shape
    # Create a canvas that can fit the above image
    w = Canvas(window, width = 300, height = 300)
    photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(cv_img))
    # Add a PhotoImage to the Canvas
    w.create_image(0, 0, image=photo, anchor=tkinter.NW)
    w.grid(column=0,row=1)
```
